```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("../datasets/UJIIndoorLoc/UJIIndoorLoc_B0-ID-01.csv")
```

Checking the data characteristics

```python
data.shape
data.describe()
data.info()
data.head(2)
```

```python
data.dtypes
```

Droping columns with all 0

```python
data = data.loc[:, (data!=0).any('index')]
data.shape
```

```python
len(data['ID'].unique())
```

###### So we for this 5249 entries we have 256 unique ID's and 201 relevant columns after droping the zero ones thus they does not give us information.
###### Therefore there we can answer the first question clarifying that would be necessary to have more entries.

```python
duplicated_data = data[data.duplicated(keep=False)]
duplicated_data.head(5)
```

Checking duplicated data of an especific ID

```python
duplicated_data_211 = duplicated_data.loc[(duplicated_data['ID']) == 211]
duplicated_data_211
```

As we can see we can have more than one kind of duplicated data per ID

```python
duplicated_data.loc[(duplicated_data['ID']) == 211].sum()
```

Example of querying with iloc function 

```python
duplicated_data_211.iloc[:, :].query('WAP007 == 1')
```

##### Dropping duplicated data

```python
data = data.drop_duplicates(keep=False)
data.shape
```

Finally we save the cleaned data set

```python
data.to_csv('../datasets/UJIIndoorLoc/UJIIndoorLoc_B0-ID-01-AfterDelivery1.csv')
```

##### Correlating data to check which data has more and less potential
Absolute value and rounding to 6 decimals

```python
corr = data.corr().abs().round(decimals=6)
corr
```

Stacking the data to manage it

```python
stacked_corr = corr.stack()
stacked_corr
```

Removing the innecessary data of the correlation (ones)

```python
stacked_corr = stacked_corr[stacked_corr.index.get_level_values(0) != stacked_corr.index.get_level_values(1)]
stacked_corr
```
